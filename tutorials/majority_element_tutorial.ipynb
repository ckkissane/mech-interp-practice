```python
import plotly.graph_objects as go

update_layout_set = {"xaxis_range", "yaxis_range", "hovermode", "xaxis_title", "yaxis_title", "colorbar", "colorscale", "coloraxis", "title_x", "bargap", "bargroupgap", "xaxis_tickformat", "yaxis_tickformat", "title_y", "legend_title_text", "xaxis_showgrid", "xaxis_gridwidth", "xaxis_gridcolor", "yaxis_showgrid", "yaxis_gridwidth"}
def imshow(tensor, renderer=None, xaxis="", yaxis="", **kwargs):
    if isinstance(tensor, list):
        tensor = torch.stack(tensor)
    kwargs_post = {k: v for k, v in kwargs.items() if k in update_layout_set}
    kwargs_pre = {k: v for k, v in kwargs.items() if k not in update_layout_set}
    if "facet_labels" in kwargs_pre:
        facet_labels = kwargs_pre.pop("facet_labels")
    else:
        facet_labels = None
    if "color_continuous_scale" not in kwargs_pre:
        kwargs_pre["color_continuous_scale"] = "RdBu"
    fig = px.imshow(utils.to_numpy(tensor), color_continuous_midpoint=0.0,labels={"x":xaxis, "y":yaxis}, **kwargs_pre).update_layout(**kwargs_post)
    if facet_labels:
        for i, label in enumerate(facet_labels):
            fig.layout.annotations[i]['text'] = label

    fig.show(renderer)

def line(tensor, renderer=None, xaxis="", yaxis="", **kwargs):
    px.line(y=utils.to_numpy(tensor), labels={"x":xaxis, "y":yaxis}, **kwargs).show(renderer)

def scatter(x, y, xaxis="", yaxis="", caxis="", renderer=None, **kwargs):
    x = utils.to_numpy(x)
    y = utils.to_numpy(y)
    px.scatter(y=y, x=x, labels={"x":xaxis, "y":yaxis, "color":caxis}, **kwargs).show(renderer)

def lines(lines_list, x=None, mode='lines', labels=None, xaxis='', yaxis='', title = '', log_y=False, hover=None, **kwargs):
    # Helper function to plot multiple lines
    if type(lines_list)==torch.Tensor:
        lines_list = [lines_list[i] for i in range(lines_list.shape[0])]
    if x is None:
        x=np.arange(len(lines_list[0]))
    fig = go.Figure(layout={'title':title})
    fig.update_xaxes(title=xaxis)
    fig.update_yaxes(title=yaxis)
    for c, line in enumerate(lines_list):
        if type(line)==torch.Tensor:
            line = utils.to_numpy(line)
        if labels is not None:
            label = labels[c]
        else:
            label = c
        fig.add_trace(go.Scatter(x=x, y=line, mode=mode, name=label, hovertext=hover, **kwargs))
    if log_y:
        fig.update_layout(yaxis_type="log")
    fig.show()

def bar(tensor, renderer=None, xaxis="", yaxis="", **kwargs):
    px.bar(
        y=utils.to_numpy(tensor),
        labels={"x": xaxis, "y": yaxis},
        template="simple_white",
        **kwargs).show(renderer)
```

# Task description

Recall the leetcode description: "Given an array nums of size n, return the majority element. The majority element is the element that appears more than $\lfloor n/2 \rfloor$ times. You may assume that the majority element always exists in the array”


The models inputs are always of the form:


[BOS, a1, a2, ..., a10, END]


Where BOS (51) and END (50) are special tokens at the start and end of every sequence. a1, ..., a10 are the array elements, which range between [0, 49] inclusive. Note that I fixed n=10 for every example during training. The model was trained to always output its prediction for the majority element at the END position.


Each example has a corresponding target, which is the majority element for that example. Below I provide the same data loader used in training, and some tokens / targets that you can use to start investigating the model:


```python
def make_data_generator(cfg, batch_size, seed=0):
    torch.manual_seed(seed)
    random.seed(seed)
    BOS_TOKEN=cfg.d_vocab-1
    END_TOKEN=cfg.d_vocab-2
    bos_vec = torch.full((batch_size, 1), BOS_TOKEN)
    end_vec = torch.full((batch_size, 1), END_TOKEN)
    majority_num = LIST_LEN // 2 + 1
    while True:
        x = torch.randint(0, MAX_ELT, (batch_size, LIST_LEN))
        majority_elt = random.randint(0, MAX_ELT-1)
        majority_indices = torch.randperm(LIST_LEN)[:majority_num]
        x[:, majority_indices] = majority_elt
        yield torch.cat([bos_vec, x, end_vec], dim=-1), torch.full((batch_size,), majority_elt)

batch_size = 256
data_loader = make_data_generator(cfg, batch_size, seed=42)

tokens, targets = next(data_loader)
tokens, targets = tokens.to(device), targets.to(device)
print("tokens:", tokens.shape)
print(tokens[:5])
print("targets:", targets.shape)
print(targets[:5])
```

I'll also provide the loss function used for training:

```python
def loss_fn(logits, labels, per_token=False):
    if logits.ndim==3:
        logits=logits[:, -1, :]
    log_probs = logits.log_softmax(dim=-1)
    correct_log_probs = log_probs.gather(dim=-1, index=labels[..., None])[..., 0]
    if per_token:
        return -correct_log_probs
    else:
        return -correct_log_probs.mean()
```

# Solution

To start, I usually find it helpful to sanity check that the model has actually learned to solve this task. This can also give us hints about what kinds of inputs the model struggles with the most.

## Sanity check

```python
with torch.inference_mode():
    logits = model(tokens)
    logits = logits[:, -1, :]
    loss = loss_fn(logits, targets)
    print("Loss:", loss.item())

    preds = logits.argmax(dim=-1)
    acc = (preds == targets).float().mean()
    print("Accuracy:", acc.item())
```

Here we can see that the model is very good at this task, with 100% accuracy and near zero loss for this sample. Normally a next step would be to localize what bits of the model seem important for this capability, but since it's only a 1H model, we can guess that that head is the most important. Let's start by staring at its attention patterns.

## Stare at attn patterns

```python
original_logits, cache = model.run_with_cache(tokens)
print(original_logits.shape)
print(cache)
```

For toy models I often find it useful to average the attn patterns over the batch dimension for some sample of the dataset. This can help us see high level trends before zooming in to individual dataset examples:

```python
def tokens_to_labels(tokens):
    if tokens.ndim == 2:
        tokens = tokens[0]
    res = [f"{tok}_{i}" for i, tok in enumerate(tokens)]
    res[0] = "BOS"
    res[-1] = "END"
    return res

labels = tokens_to_labels(tokens)
print(labels)
```

```python
layer=0
head_index=0
cv.attention.attention_patterns(
    cache['pattern', layer].mean(dim=0),
    tokens=labels,
    attention_head_names=["0.0"]
)
```

We see that END attends roughly uniformly to all of the non special tokens. Let's see if this holds true for individual examples:

```python
all_labels = [tokens_to_labels(tok) for tok in tokens]

cv.attention.from_cache(
    cache = cache,
    tokens = all_labels,
)
```

"END attends uniformly to all non special tokens" holds true for every example that I checked. However all of the inputs in this sample are quite similar, so we could be missing edge cases that don't follow this pattern.


One way to gain a more general understanding is to look directly at the weights. Since this model is so simple, we can write attn scores for this head as $\frac{tW_EW_{QK}^{0.0}W_E^Tt^T}{\sqrt{d_{head}}}$, where $t$ is the input as one hot encoded vectors. (If this isn't clear, please reference [A Mathematical Framework for Transformer Circuits](https://transformer-circuits.pub/2021/framework/index.html) or my [min of two ints tutorial](https://colab.research.google.com/github/ckkissane/mech-interp-practice/blob/main/tutorials/min_of_two_ints_tutorial.ipynb#scrollTo=I2dv4B_eWwyl).) Thus we should be able to interpret the "full QK circuit", $W_EW_{QK}^{0.0}W_E^T$, a lookup table which maps each possible (dest token, src token) pair to a corresponding attention score (before scaling).

## Stare at full QK circuit

```python
W_E = model.W_E
print(W_E.shape)
```

```python
token_labels = list(map(str, range(cfg.d_vocab)))
token_labels[-1] = "BOS"
token_labels[-2] = "END"
```

```python
layer, head_index = 0, 0
imshow(
    W_E @ model.QK[layer,head_index].AB @ W_E.T,
    title=f"{layer}.{head_index} Full QK circuit",
    xaxis="src token", yaxis="dest token",
    x=token_labels, y=token_labels
)

bar(
    (W_E @ model.QK[layer,head_index].AB @ W_E.T)[-2],
    title=f"{layer}.{head_index} Full QK circuit (for dest END token)",
    xaxis="src token", yaxis="attn score",
    x=token_labels,
)
```

Since we only really care about destination token END (since the prediction is made at this position), I also visualize that row as a bar graph. Notice that END does indeed give roughly uniform attn scores to all non special tokens, as expected.


Now that we understand how 0.0 attends roughly uniformly to all the array elements, how does it use this to predict the majority element?


Recall that we can write the contribution of this head to the logits as $A^{0.0}tW_EW_{OV}^{0.0}W_U$ where $A^{0.0}$ is the attention pattern for 0.0 and $t$ is the input tokens as one hot encoded vectors. Thus we should be able to interpret the full OV circuit, $W_EW_{OV}^{0.0}W_U$, a lookup table which maps "src token that we attend to" to the corresponding effect on the logits.

## Stare at full OV circuit

```python
W_U = model.W_U
print(W_U.shape)
```

```python
imshow(
    W_E @ model.OV[layer, head_index].AB @ W_U,
    title=f"{layer}.{head_index} full OV circuit",
    xaxis="output token logits", yaxis="src token"
)
```

We see a clear copying OV matrix: "If I attend to token X, then boost Xs logits". This makes sense: Uniform attention will always pay the most aggregate attention to the majority element, so the OV circuit can just copy to predict that token.

## Wrap up / Summary

Mechanistic interpretability is often easy to start, but hard to finish. Although we have a high level understanding of the weights, I can't claim to fully understand every single entry in the QK / OV matrices, so it's possible that there are adversarial examples that could exploit this lack of understanding. However we were able to quickly 80/20 an understanding of how this model solved the majority element problem using just two techniques:


1. Staring at attention patterns
2. Multiplying out QK / OV circuits

We also found another concrete example of a "copying" OV circuit, a common motif that has appeared in circuits such as [induction](https://transformer-circuits.pub/2022/in-context-learning-and-induction-heads/index.html#argument-induction-head-mechanism), [IOI](https://arxiv.org/abs/2211.00593), and many other toy tasks.
